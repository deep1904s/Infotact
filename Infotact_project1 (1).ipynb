# Feature 1: Match Outcome Analysis

```python
team_wins = matches['winning_team'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values, palette="viridis")
plt.xticks(rotation=90)
plt.title("Number of Matches Won by Each Team")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.show()
```

```python
matches_per_season = matches['season'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette="viridis")
plt.title("Total Matches Played Across Seasons (2008-2022)")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()

team_wins_per_season = matches.groupby(['season', 'winning_team']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
team_wins_per_season.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14, 8))
plt.title("Matches Won by Teams Across Seasons")
plt.xlabel("Season")
plt.ylabel("Number of Wins")
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

margin_per_season = matches.groupby('season').mean()
```

# Feature 2: Player Performance

```python
def player_performance(player_name):
    player_data = data[data['batter'] == player_name]
    player_runs = player_data['batsman_run'].sum()
    player_balls = player_data.shape[0]
    strike_rate = (player_runs / player_balls) * 100
    print(f"{player_name}'s Total Runs: {player_runs}")
    print(f"{player_name}'s Strike Rate: {strike_rate:.2f}")

    
if __name__ == "__main__":
    print("\nFeature 2: Player Performance")
    player_name = "V Kohli" 
    player_performance(player_name)
```

```python
batsman_runs = deliveries.groupby('batter')['batsman_run'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=batsman_runs.values, y=batsman_runs.index, palette="Blues_d")
plt.title("Top 10 Batsmen by Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
pom_counts = matches['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=pom_counts.values, y=pom_counts.index, palette="cubehelix")
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Number of Awards")
plt.ylabel("Players")
plt.show()
```

# Feature 3: Team Comparison

```python
matches_played = matches['team1'].value_counts() + matches['team2'].value_counts()
win_percentage = (team_wins / matches_played * 100).sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=win_percentage.index, y=win_percentage.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Team Win Percentage (2008-2022)")
plt.xlabel("Teams")
plt.ylabel("Win Percentage (%)")
plt.show()
```

# Feature 4: Venue Performance 

```python
venue_counts = matches['venue'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette="crest")
plt.title("Number of Matches Played at Each Venue")
plt.xlabel("Number of Matches")
plt.ylabel("Venues")
plt.show()
```

# Feature 5: Run Rate & Scoring Analysis

```python
def run_rate_analysis():
    deliveries['run_rate'] = deliveries.groupby('id')['total_run'].cumsum() / deliveries['overs']
    match_run_rate = deliveries.groupby('overs')['run_rate'].mean()
    match_run_rate.plot(kind='line', figsize=(12, 6), color='purple')
    plt.title('Average Run Rate Over Overs', fontsize=16)
    plt.xlabel('Overs')
    plt.ylabel('Run Rate')
    plt.show()
if __name__ == "__main__":
    
    print("\nFeature 5: Run Rate & Scoring Analysis")
    run_rate_analysis()
```
